```python
def carrier_curtailed(carrier):
    
    result = {}
    
    carrier_generators = network.generators.carrier[network.generators.carrier == carrier].index.tolist()

    p_by_bus = network.generators_t.p[carrier_generators].groupby(
        network.generators.bus, axis=1).sum()

    capacity = network.generators.groupby("carrier").sum().at[carrier, "p_nom"]
    p_available = network.generators_t.p_max_pu.multiply(network.generators["p_nom"])[carrier_generators]

    p_available_by_bus = p_available.groupby(network.generators.bus, axis=1).sum()

    p_curtailed_by_bus = p_available_by_bus - p_by_bus
    
    result['available'] = p_available_by_bus.sum().sum()/1e6
    
    result['curtailed'] = p_curtailed_by_bus.sum().sum()/1e6
    
    print(carrier)
    print(result)
    

    return p_curtailed_by_bus
```
